## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import datetime
test_date = datetime.datetime(2023, 1, 1)
K = 40
res = [test_date + datetime.timedelta(days=idx) for idx in range(K)]
print("Next K dates list : " + str(res))
```
